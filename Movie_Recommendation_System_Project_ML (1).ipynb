```python
vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_data_set)
```

***Checking the Similarity***

```python
Similarity=cosine_similarity(feature_vector)
```

***Making The Predictive System***

```python
print("Welcome To Movie Recommendation System!")
print()
user_input=input("Enter Your Favoutite Show!")
Find_close_match=difflib.get_close_matches(user_input,Movie_List)


Movie_List=data_set["original_title"].tolist()



close_match=Find_close_match[0]

filtered_data = data_set[data_set.title == close_match]


if not filtered_data.empty:
    close_index = filtered_data["index"].values[0]
    print(f"Index found: {close_index}")
else:
    print("Movie not found in dataset!")


similarity_score=list(enumerate(Similarity[close_index]))


sorting_similarity=sorted(similarity_score ,key=lambda x:x[1],reverse=True)


print("Here are the suggested movie for you : \n")


i=0

for k in sorting_similarity:
  index=k[0]
  title_from_index=data_set[data_set.index==index]["title"].values[0]

  if i<30:
    print(i,".",title_from_index)
    i+=1
```
